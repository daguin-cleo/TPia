### **I) XOR gate**
Nous allons d'abord commencer par étudier les éléments de base d'un réseau de neurones à travers un réseau de neurones sur les portes XOR. Source : xor.py fournie avec le tp

Imports nécessaire pour notre TP

```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD
from matplotlib import pyplot
from keras.utils import plot_model
import numpy as np
import warnings
```

*   Pour faire fonctionner un réseau de neurons, il faut le faire travailler sur un **set de données d'entrainement**. Pour cela nous avons construit X et Y respectivement l'entrée et la sortie voulue du réseau. Nous avons donc Y = X1 XOR X2
*   model est notre **réseau de neurons**.
*   Pour créer notre réseau de neurons, nous devons lui ajouter des **couches successives** (d'où Sequential())
    *   La première couche sera une couche de 8 neurones qui prennent tous en entrée les 2 dimensions des données d'entrée (X1 et X2)
    *   Ensuite vient une couche avec un seul neurone qui est la couche de sortie. Elle est activée grâce à la fonction sigmoid qui va transformer le résultat en un format de sortie souhaitée, soit 0 ou 1.
*   **compile** crée le réseau à partir des caractéristiques des différentes couches que nous avons choisit. 
    *   Le paramètre **loss** va permettre de voir si les sorties se rapprochent ou pas des sorties attendues. Ici, nous utilisons la "binary_crossentropy" qui va prendre en compte non seulement que les sorties attendues à 1 soient à 1 mais aussi que les autres sorties soient à 0. Plus la loss est basse, plus les sorties vont etre proches des résultats attendus. 
    *   Le paramètre **optimizer** va permettre d'avoir un meilleur résultat via différentes méthodes en fonction de l'optimiseur. Ici sgd va enlever les valeurs abherrantes. 
*   **fit** va entrainer le réseau sur les données de train. Pour l'entrainer on ne fait pas tout le set de données d'entrée d'un coup mais un certain nombre d'échantillons plusieurs fois de suite avec un changement des échantillons à chaque itération.
    *   **X et Y** sont les entrées et les sorties attendues.
    *   **batch_size** est le nombre d'échantillons en entrée à chaque itération du model. Ici, comme le set de données d'entrée est très restreint, nous n'allons faire des itérations qu'avec un échantillon à la fois. 
    *   **epoch** est le nombre d'itérations sur le réseau de neurones


```python
warnings.filterwarnings('ignore')
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

model = Sequential()
model.add(Dense(8, input_dim=2))
model.add(Activation('tanh'))
model.add(Dense(1))
model.add(Activation('sigmoid'))

sgd = SGD(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer=sgd)

model.fit(X, y, batch_size=1, epochs=1000)
print(model.predict_proba(X))
```

Affichage de la taille du réseau de neurones et de ses différentes couches. 

```python
plot_model(model, to_file='model1.png', show_shapes='True')
```

Maintenant, nous allons construire un réseau de neurones qui avec en entrée un vecteur de binaires à 2 dimensions nous donne en sortie **xor, and, or, notX1 et notX2** en même temps. Pour cela nous devons construire les sorties attendues.

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
#xor
y1 = np.array([[0],[1],[1],[0]])
#and
y2 = np.array([[0],[0],[0],[1]])
#or
y3 = np.array([[0],[1],[1],[1]])
#not x_1
y4 = np.array([[1],[1],[0],[0]])
#not x_2
y5 = np.array([[1],[0],[1],[0]])
#concatenation des précédents
y6 = np.concatenate((y1,y2,y3,y4,y5),axis=1)
```

Pour le réseau, il est construit comme le précédent mais nous allons avoir 5 sorties pour le résultat des 5 portes différentes.

```python
model = Sequential()
model.add(Dense(8, input_dim=2))
model.add(Activation('tanh'))
model.add(Dense(5))
model.add(Activation('sigmoid'))

sgd = SGD(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer=sgd)

model.fit(X, y6, batch_size=1, epochs=1000)
print(model.predict_proba(X))
```

Le paramètre epochs va indiquer le nombre d'itérations que le réseau fait sur les données. Plus il y a d'epochs plus le résultat obtenu sera proche du résultat attendu. Il n'y a pas de valeur optimal, c'est un équilibre entre le temps de calcul qui augment avec chaque epoch rajouté et la précision recherchée, sachant que l'obtention de résultats valides à 100% est souvent inatteignable.


Il est possible d'ajouter une couche cachée de neurones plus petite entre les deux autres mais le résultat peu être moins précis.

### **II) MNIST dataset**
Nous allons appliquer ce que nous avons appris plus haut pour comprendre un réseau neuronal fourni qui identifie un chiffre écrit à la main. Source : https://github.com/keras-team/keras/blob/master/examples/mnist_mlp.py

```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop
```

**num_classes** est le nombre de sorties. Ici il y a 10 sorties car il y a 10 chiffres possibles. La sortie globale sera donc un vecteur binaire de longueur 10 et qui a tous ses éléments à 0 sauf un à 1 qui aura pour indice le chiffre identifié.

```python
batch_size = 128
num_classes = 10
epochs = 20
```

**Téléchargement des données et séparation** entre les données d'entrainement (train) et de test (test) et entre les entrées (x) et les sorties(y). Il est important d'avoir un set de données pour l'entrainement et un set de données différent pour le test car cela permet de voir si le réseau de neurons est bien entrainé.

```python
(x_train, y_train),(x_test, y_test)=mnist.load_data()
```

```python
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')

# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

Pour bien comprendre ce réseau, nous avons regardé les sorties attendues. C'est ainsi que nous avons compris le vecteur de taille 10 avec un seul élément à 1. Ici, c'est une image du chiffre 5.

```python
print(y_train[0])
```

Nous avons fait la somme de tous les vecteurs de sortie et nous avons le nombre d'images montrant chaque chiffre : dans notre dataset, il y a 5923 images de 0 par exemple.

```python
print(np.sum(y_train,axis=0))
```

Nous avons repris le code fourni et nous l'avons exécuté.

```python


model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(784,)))
model.add(Dropout(0.2))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(num_classes, activation='softmax'))

model.summary()

model.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])

history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(x_test, y_test))
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])

```

Nous avons ensuite fais quelques modifications pour en voir les conséquences.

Ici, nous avons diminuer le nombre de neurones par couche. Nous pouvons voir que l'accuracy est un peu moins bonne. Par contre, le temps de calcul a considérablement baissé en passant de 8s par epoch à 1s. On peut donc conclure qu'avec moins de neurones, le réseau est plus rapide mais moins performant ce qui est logique car le calcul est moins fin.

```python
model = Sequential()
model.add(Dense(40, activation='relu', input_shape=(784,)))
model.add(Dropout(0.2))
model.add(Dense(40, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(num_classes, activation='softmax'))

model.summary()

model.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])

history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(x_test, y_test))
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

Ici nous avons modifier la fonction d'activation des couches d'entrée et intermédiaire. on peut voir que l'accuracy est presque pareil mais que la loss avec tanh est bien meilleur. Les fonctions d'activations sont importantes mais pour choisir laquelle utiliser, cela dépend du réseau.

```python
model = Sequential()
model.add(Dense(512, activation='tanh', input_shape=(784,)))
model.add(Dropout(0.2))
model.add(Dense(512, activation='tanh'))
model.add(Dropout(0.2))
model.add(Dense(num_classes, activation='softmax'))

model.summary()

model.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])

history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(x_test, y_test))
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

Ici, nous avons augmenter le dropout. L'accuracy est un peu moins importante mais la val-loss est meilleur et n'augmente pas autant au bout d'un certain nombre d'itérations. Le dropout met certains poids à 0 et permet d'éviter l'overfitting mais est plus lent à converger vers une bonne précision.

```python
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(784,)))
model.add(Dropout(0.7))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.7))
model.add(Dense(num_classes, activation='softmax'))

model.summary()

model.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])

history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(x_test, y_test))
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

#### **III) IMDB dataset**
Cette partie a pour but de comprendre l'overfitting et l'underfitting. Pour ce faire, nous utiliserons une base de données qui à des commentaires sur des films indique si ils sont positifs ou négatifs. Source : https://github.com/tensorflow/docs/blob/master/site/en/r1/tutorials/keras/overfit_and_underfit.ipynb

```python


from __future__ import absolute_import, division, print_function, unicode_literals

try:
  # %tensorflow_version only exists in Colab.
  %tensorflow_version 2.x
except Exception:
  pass
import tensorflow.compat.v1 as tf

from tensorflow import keras

import numpy as np
import matplotlib.pyplot as plt

print(tf.__version__)

```

Les données téléchargées sont sous la forme d'une liste d'indices des mots du commentaire et d'un chiffre binaire qui indique si le commentaire est positif ou négatif. Les mots sont enregistrés dans un dictionnaire et les indices correspondent au classement des mots en fonction de leur fréquence : le mot avec l'indice 1 est le mot le plus courant. NUM_WORDS va déterminer combien de mots différents, le réseau de neurons va prendre en compte. Ici,il prend en compte les 10000 mots les plus courants et donne aux autres une valeur neutre. Cela diminue les calculs.De plus, moins un mots est courant moins il a d'intérêt.

```python

NUM_WORDS = 10000

(train_data, train_labels), (test_data, test_labels) = keras.datasets.imdb.load_data(num_words=NUM_WORDS)
```

Pour nous familiariser avec les données nous avons fait quelques manipulations dessus. Nous pouvons voir pour chaque film les mots de son commentaire représentés par leur indice dans le dictionaire.

```python
train_data
```

Ici on peut voir les mots travers leur indice dans le dictionnaire en fonction de leur place dans le commentaire. On observe bien qu'il y a plus de mots avec un indice proche de 0 que de mots avec un indice autour de 8000.

```python
plt.plot(train_data[2])
```

Ensuite nous avons reconstitué un commentaire à partir du vecteur d'indices et du dictionnaire.

```python
INDEX_FROM=3

word_to_id = keras.datasets.imdb.get_word_index()
word_to_id = {k:(v+INDEX_FROM) for k,v in word_to_id.items()}
word_to_id["<PAD>"] = 0
word_to_id["<START>"] = 1
word_to_id["<UNK>"] = 2
word_to_id["<UNUSED>"] = 3

id_to_word = {value:key for key,value in word_to_id.items()}
print(' '.join(id_to_word[id] for id in train_data[400] ))
```

Les données récupérées sont sous forme de liste de mots. Il faut donc en faire des **vecteurs** de taille NUM_WORDS et mettre 1 si le mot est présent dans le commentaire et 0 sinon. Ainsi toutes les entrées auront le **même format**.

```python
def multi_hot_sequences(sequences, dimension):
    # Create an all-zero matrix of shape (len(sequences), dimension)
    results = np.zeros((len(sequences), dimension))
    for i, word_indices in enumerate(sequences):
        results[i, word_indices] = 1.0  # set specific indices of results[i] to 1s
    return results


train_data = multi_hot_sequences(train_data, dimension=NUM_WORDS)
test_data = multi_hot_sequences(test_data, dimension=NUM_WORDS)
```

```python
plt.plot(train_data[0])
```

Nous créons un model pour déterminer si les commentaires sont positifs ou négatifs.

```python
baseline_model = keras.Sequential([
    keras.layers.Dense(16, activation=tf.nn.relu, input_shape=(NUM_WORDS,)),
    keras.layers.Dense(16, activation=tf.nn.relu),
    keras.layers.Dense(1, activation=tf.nn.sigmoid)
])

baseline_model.compile(optimizer='adam',
                       loss='binary_crossentropy',
                       metrics=['accuracy', 'binary_crossentropy'])

baseline_model.summary()
```

```python
baseline_history = baseline_model.fit(train_data,
                                      train_labels,
                                      epochs=20,
                                      batch_size=512,
                                      validation_data=(test_data, test_labels),
                                      verbose=2)
```

Maintenant nous utilisons un **model plus petit**

```python
smaller_model = keras.Sequential([
    keras.layers.Dense(4, activation=tf.nn.relu, input_shape=(NUM_WORDS,)),
    keras.layers.Dense(4, activation=tf.nn.relu),
    keras.layers.Dense(1, activation=tf.nn.sigmoid)
])

smaller_model.compile(optimizer='adam',
                loss='binary_crossentropy',
                metrics=['accuracy', 'binary_crossentropy'])

smaller_model.summary()
```

```python
smaller_history = smaller_model.fit(train_data,
                                    train_labels,
                                    epochs=20,
                                    batch_size=512,
                                    validation_data=(test_data, test_labels),
                                    verbose=2)
```

```python
bigger_model = keras.models.Sequential([
    keras.layers.Dense(512, activation=tf.nn.relu, input_shape=(NUM_WORDS,)),
    keras.layers.Dense(512, activation=tf.nn.relu),
    keras.layers.Dense(1, activation=tf.nn.sigmoid)
])

bigger_model.compile(optimizer='adam',
                     loss='binary_crossentropy',
                     metrics=['accuracy','binary_crossentropy'])

bigger_model.summary()
```

```python
bigger_history = bigger_model.fit(train_data, train_labels,
                                  epochs=20,
                                  batch_size=512,
                                  validation_data=(test_data, test_labels),
                                  verbose=2)
```

```python
def plot_history(histories, key='binary_crossentropy'):
  plt.figure(figsize=(16,10))

  for name, history in histories:
    val = plt.plot(history.epoch, history.history['val_'+key],
                   '--', label=name.title()+' Val')
    plt.plot(history.epoch, history.history[key], color=val[0].get_color(),
             label=name.title()+' Train')

  plt.xlabel('Epochs')
  plt.ylabel(key.replace('_',' ').title())
  plt.legend()

  plt.xlim([0,max(history.epoch)])


plot_history([('baseline', baseline_history),
              ('smaller', smaller_history),
              ('bigger', bigger_history)])
```

On peut observer que le model de base et le model réduit ont des epochs court par rapport au grand model. C'est normal car le grand model a beaucoup plus de neurons donc beaucoup plus de calculs à faire.
Pour la **val-loss**, le model de base a une val-loss qui diminue jusqu'au 2eme epoch et qui ensuite ne fait qu'augmenter. Le réduit a une val-loss qui diminue jusqu'au 5 eme epoch et le grand a une val-loss qui augment instantanément. Or, l'augmentation de laval-loss montre un overfitting. C'est à dire que le modèle colle aux données d'entrainement et ne prend pas en compte la logique générale. Lorsque le model sera utilisé sur un autre set de données, les résultats ne seront pas aussi bons que ceux de l'entrainement. On le voit plus sur un grand model car il stocke chaque détail des données alors qu'avec un plus petit modèle ce n'est pas possible et le modèle doit donc s'appuyer sur un comportement plus général.
